```python
# Task: Image Captioning
captioner = pipeline("image-to-text",
                     model = "nlpconnect/vit-gpt2-image-captioning")
```

```python
captioner2 = pipeline("image-to-text",
                      model = "microsoft/git-base")
```

```python
captioner3 = pipeline("image-to-text",
                      model = "Salesforce/blip-image-captioning-base")
```

```python
images = Image.open('Football.jpg')
display(Image.open('Football.jpg'))
print(captioner('Football.jpg'))
print(captioner2('Football.jpg'))
print(captioner3('Football.jpg'))
```
